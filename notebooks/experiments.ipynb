```python
from binance.client import Client
import os
```

```python
api_key = os.getenv('API_KEY')
secret_key = os.getenv('SECRET_KEY')
```

```python
api_key
```

```python
client = Client(api_key, secret_key)
client.FUTURES_URL = "https://testnet.binancefuture.com/fapi"

```

```python
symbol = "BTCUSDT"
side = "SELL"
order_type = "MARKET"
quantity = 0.01
price = None

response = client.futures_create_order(
    symbol=symbol,
    side=side,
    type=order_type,
    quantity=quantity,
    price=price,
    # timeInForce="GTC"
)

print(response)
```

```python
symbol = "BTCUSDT"
side = "BUY"
order_type = "STOP"

quantity = 0.01
stop_price = 50000
price = 50100

response = client.futures_create_order(
    symbol=symbol,
    side=side,
    type=order_type,
    quantity=quantity,
    stopPrice=stop_price,
    price=price,
    timeInForce="GTC"
)

print(response)
```

```python
ticker = client.futures_symbol_ticker(symbol="BTCUSDT")
print("Current price:", ticker["price"])
```
